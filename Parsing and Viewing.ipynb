```python
from Bio import SeqIO
import pandas as pd

def get_a_view(filename) :

    parsed_data = SeqIO.parse(filename, "fasta")
    counter = 0
    unique_id = set()
    data = []
    
    for each_data in parsed_data :
        
        id = each_data.id.split("/")[-1].split('fq')[0][:-1]
        seq = str(each_data.seq)
        
        data.append([id,seq])
        unique_id.add(id)
        
    df = pd.DataFrame(data , columns = ["Accession_ID" , "250bp_READ"])

    return(unique_id , df)
```

```python
ids, dataframe = get_a_view("datasets/pathogenic_val.fasta")
```
